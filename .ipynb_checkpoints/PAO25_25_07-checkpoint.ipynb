#  Pandas

* Librería (de facto estándar) para estructurar datos tabulares
* Multivariable (string, int, float, bool…)
* Dos clases:
  
    – Series (1 dimensión)
  
    – DataFrames (2+ dimensiones)



```python
# librería externa
import pandas as pd
from pandas import Series, DataFrame
```

```python
x = 3
y = [1, 2, 3]
print(x > 0)
print(x > 0 and x < 10)
print(x is not y)
print(x in y)
```

* Se pueden introducir más ‘ramas’ en sentencias condicionales a través de la palabra clave elif.


```python
x = -3
if x > 0:
    print('Valor positivo')
elif x == 0:
    print('Valor nulo')
else:
    print('Valor negativo')
```

* Anidamiento.

```python
val = 120
if val > 0:
  if val < 100:
    print('Valor positivo')
  else:
    print('Valor muy positivo')
else:
    print('Valor negativo')
```

# Switch

```python
a = -6
match a:
    case _ if a > 0:
      print('positive')
    case _ if a == 0:
      print('zero')
    case _ if a < 0:
      print('negative')
```

```python
a = 'l' #week day
match a:
    case 'L':
     print("Lunes")
    case 'M':
     print("Martes")
    case 'X':
     print("Miercoles")
    case _:
     print("Wrong Day")
```
